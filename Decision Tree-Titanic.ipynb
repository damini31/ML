```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
```

```python
df=pd.read_excel("titanic.xlsx")
```

```python
df.head()
```

```python
X=df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Survived'],axis='columns')
```

```python
X
```

```python
Y=df.Survived
```

```python
Sex_n=LabelEncoder()
Embarked_n=LabelEncoder()
```

```python
X['Sex_new']=Sex_n.fit_transform(X['Sex'])
```

```python
X['Embarked'] = X['Embarked'].fillna('')
X['Embarked_new']=Embarked_n.fit_transform(X['Embarked'])
```

```python
X
```

```python
X_new=X.drop(['Sex','Embarked'],axis="columns")
```

```python
X_new
```

```python
model=tree.DecisionTreeClassifier()
```

```python
X_new['Age']=X_new['Age'].fillna(0.0)
```

```python
X_new
```

```python
X_new.dtypes
```

```python
model.fit(X_new,Y)
```

```python
model.score(X_new,Y)
```
